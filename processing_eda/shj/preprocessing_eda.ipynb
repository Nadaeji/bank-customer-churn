```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 한글 설정
import matplotlib.font_manager as fm

# 맑은고딕 경로
font_path = "C:\\Windows\\Fonts\\malgun.ttf"

font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
```

### 데이터 로드

```python
# 데이터 로드
bank_df = pd.read_csv('../../data/BankChurners.csv')
display(bank_df)
```

```python
display(bank_df.describe())
display(bank_df.info())
```

```python
# 이탈여부 값 변환
bank_df['Attrition_Flag'] = bank_df['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})

# 은행 관계와 관련된 칼럼만 추출
bank_rel_columns = ['Months_on_book', 'Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon']
bank_rel_df = bank_df[bank_rel_columns + ['Attrition_Flag']]

# 한글 번역
bank_rel_columns_kor_dict = {
    'Months_on_book' : '가입 기간(개월)',
    'Total_Relationship_Count' : '보유한 상품의 수',
    'Months_Inactive_12_mon' : '지난 12개월 동안 거래를 하지 않은 기간(개월)',
    'Contacts_Count_12_mon' : '지난 12개월 동안 연락 횟수'
}

display(bank_rel_df.describe())
```

### 피어슨 상관계수 확인 및 히트맵 표시

```python
# 피어슨 상관계수
# 이탈 변수 관련한 상관계수만만 추출
bank_rel_corr = bank_rel_df.corr()
churn_corr = bank_rel_corr['Attrition_Flag'].sort_values(ascending=False)
print(churn_corr)

# 히트맵 표시
plt.figure(figsize=(6, 6))
sns.heatmap(bank_rel_df.corr(), annot=True)
plt.title('Correlation Heatmap of Bank Relationship Features')
plt.show()
```

### 은행 가입 고객 분포 확인

```python
# 은행 가입 고객 분포 확인
plt.figure(figsize=(8, 8))
plt.pie(bank_rel_df['Attrition_Flag'].value_counts(), labels=['기존 고객', '이탈 고객'], autopct='%1.1f%%', startangle=0, explode=[0.1, 0], wedgeprops={'width': 0.75}, colors=['skyblue', 'orange'])
plt.title('은행 가입 고객 분포')
plt.show()
```

```python
# 기존 고객과 이탈 고객 분리
bank_exist_df = bank_rel_df[bank_rel_df['Attrition_Flag'] == 0]
bank_churn_df = bank_rel_df[bank_rel_df['Attrition_Flag'] == 1]
```

### 유지 고객과 이탈 고객의 각 Feature 별 분포

##### 비율 분포

```python
# bar + plot 차트
plt.figure(figsize=(16, 16))

for i, col in enumerate(bank_rel_columns, 1):
    plt.subplot(2, 2, i)
        
    # 비율 계산
    exist_ratio = bank_exist_df[col].value_counts(normalize=True).sort_index() * 100
    churned_ratio = bank_churn_df[col].value_counts(normalize=True).sort_index() * 100
    
    # 막대그래프 생성
    width = 0.3
    plt.bar(exist_ratio.index - width/2, exist_ratio.values, width=width, label='유지 고객', color='blue', alpha=0.3)
    plt.bar(churned_ratio.index + width/2, churned_ratio.values, width=width, label='이탈 고객', color='orange', alpha=0.3)
    
    # KDE 플롯 추가
    plt.plot(exist_ratio.index, exist_ratio.values, marker='o', color='blue')
    plt.plot(churned_ratio.index, churned_ratio.values, marker='o', color='orange')
    
    # 그래프 설정
    plt.title(f'{bank_rel_columns_kor_dict[col]} 분포')
    plt.xlabel(bank_rel_columns_kor_dict[col])
    plt.ylabel('Percentage (%)')
    plt.legend()

plt.tight_layout()
plt.show()
```

##### 밀집도 분포

```python
# 밀집도 분포 확인
plt.figure(figsize=(16, 12))
for i, col in enumerate(bank_rel_columns, 1):
    plt.subplot(2, 2, i)
    sns.kdeplot(bank_exist_df[col], label='유지 고객', shade=True, color='blue')
    sns.kdeplot(bank_churn_df[col], label='이탈 고객', shade=True, color='orange')
    plt.title(f'{bank_rel_columns_kor_dict[col]} 분포')
    plt.xlabel(f'{bank_rel_columns_kor_dict[col]}')
    plt.legend()

plt.tight_layout()
plt.show()
```
